```python
package rest_assured_assignment;

import com.aventstack.extentreports.ExtentReports;
import com.aventstack.extentreports.ExtentTest;
import com.aventstack.extentreports.reporter.ExtentSparkReporter;
import com.aventstack.extentreports.reporter.configuration.Theme;
import io.restassured.http.ContentType;
import io.restassured.response.Response;
import org.testng.Assert;
import org.testng.annotations.AfterClass;
import org.testng.annotations.BeforeClass;
import org.testng.annotations.Test;

import static io.restassured.RestAssured.given;

public class Assignment_001 {
    private ExtentSparkReporter spark;
    private ExtentReports extent;
    private ExtentTest logger;

    public Assignment_001_RequestBody req_object;

    @BeforeClass
    public void setup()
    {
        req_object = new Assignment_001_RequestBody();
        // C:\Users\maheswaranma\IdeaProjects\EdYoda08_TestNG
        extent = new ExtentReports();
        spark = new ExtentSparkReporter(System.getProperty("user.dir") + "/TestReport/Assignment_001.html");
        extent.attachReporter(spark);
        extent.setSystemInfo("Environment","QA");
        extent.setSystemInfo("App Version","1.0");
        extent.setSystemInfo("UserName","Mahesh");
        spark.config().setDocumentTitle("Doing testing for Assignment_001");
        spark.config().setTheme(Theme.DARK);
        logger = extent.createTest("Creating test report for Assignment_001");
    }

    @Test
    public void pet_id_E2E_testing()
    {
        for(int i=200;i<220;i++)
        {
            String id = String.valueOf(i);
            String name = "Animal_"+id;
            pet_id_creation(id,name);
            retrieve_pet_id(id);
            delete_pet_id(id);
        }

    }
    public void pet_id_creation(String pet_id, String pet_name)
    {
        Response res = given()
                .contentType(ContentType.JSON)
                .body(req_object.create_pet_RequestBody_001(pet_id, pet_name))
                .when()
                .post("https://petstore.swagger.io/v2/pet");
        int create_post_status_code = res.getStatusCode();
        System.out.println("The creation status code = " +create_post_status_code);
        String pet_id_res = res.getBody().jsonPath().getString("id");
        String pet_name_res = res.getBody().jsonPath().getString("name");
        System.out.println("The pet id is "+pet_id_res);
        System.out.println("The pet name is "+pet_name_res);
        logger.pass("The status code is "+create_post_status_code);
        logger.info("The pet id is "+pet_id);
        logger.info("The pet name is "+pet_name);
    }

    public void retrieve_pet_id(String pet_id)
    {
        String runtime_pet_id = "https://petstore.swagger.io/v2/pet/"+pet_id;
        Response res = given()
                .when()
                .get(runtime_pet_id);
        int get_status_code = res.getStatusCode();
        System.out.println("The retrieve status code = " +get_status_code);
        Assert.assertTrue(get_status_code==200);
        logger.pass("The retrieve status code is "+get_status_code);

    }

    public void delete_pet_id(String pet_id)
    {
        String runtime_pet_id = "https://petstore.swagger.io/v2/pet/"+pet_id;
        Response res = given()
                .when()
                .delete(runtime_pet_id);
        int delete_status_code = res.getStatusCode();
        System.out.println("The delete status code = " +delete_status_code);
        Assert.assertTrue(delete_status_code==200);
        logger.pass("The delete status code is "+delete_status_code);
    }

    @AfterClass
    public void tearDump()
    {
        extent.flush();
    }

}
```
